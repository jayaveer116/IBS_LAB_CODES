```python
# atomic weights
atomic_weights= {'C': 12.01, 'H':1.01, 'N':14.01, 'O': 16.00}

# Molecular formula of glycine
molecular_formula = {'C': 2, 'H':5, 'N':1, 'O': 2}

# Calculate Molecular weight
molecular_weight = sum(atomic_weights[element]*count for element,count in molecular_formula.items())

print(f"The molecular weight of glyice is {molecular_weight:.2f} g/mol")

# Antibiotic Sequencing: Molecular weight using Biopython module for amino acids
```
